```python
plt.figure(figsize=(25,50))
plotnumber = 1

for column in data1:
    ax = plt.subplot(6,2,plotnumber)
    sns.countplot(data, x=data1[column])
    plt.xlabel(column, fontsize=20)
    plt.ylabel('count', fontsize=20)
    plotnumber+=1
plt.tight_layout()
```
